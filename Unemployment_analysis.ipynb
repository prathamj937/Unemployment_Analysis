```python
import pandas as pd
import numpy as np
import calendar
```

```python
df = pd.read_csv("/content/Unemployment_Rate_upto_11_2020 - Unemployment_Rate_upto_11_2020.csv")
df.head()
```

Since all dtype are correct for respective columns therefore we are not changing anything


```python
df.info()
```

Checking any null values

```python
df.isnull().sum() # No NUll Values
```

Rename some columns for better understanding

```python
df = df.rename(columns={'Region': 'State'})
df = df.rename(columns={'Region.1': 'Location'})
```

```python
df.head()
```

Converting 'Frequency' and 'Region' columns to categorical data type

```python
df['Frequency'] = df['Frequency'].astype('category')
df['Location'] = df['Location'].astype('category') # improves efficiency of the code
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
```

```python
month_map = {1: 'January', 2: 'February', 3: 'March', 4: 'April',
                 5: 'May', 6: 'June', 7: 'July', 8: 'August',
                 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
df['Month_Name'] = df['Month'].apply(lambda x: month_map.get(x, 'Invalid Month'))
```

```python
df.head()
```

```python
df.columns
```

Just doing some statistical work to get a sneek peek of the dataset

```python
df_stat = df[['Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']]
print(round(df_stat.describe().T, 1))
```

```python
region_stats = df.groupby(['State'])[['Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']].mean().reset_index().sort_values(by='Estimated Unemployment Rate (%)', ascending=False)
print(round(region_stats))
```

Now I have made some data visualization and provide some insights through each visualization(NOTE : I have made insights till 10th month)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Month', y='Estimated Unemployment Rate (%)', hue='State')
plt.title('Regional Unemployment Trends')
plt.xlabel('Month')
plt.ylabel('Unemployment Rate (%)')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(True)
plt.tight_layout()
plt.show()
```

1.) From the above Bar Graph Represnetation we can tell that During April and May Unemployment Rate reaches its peak.

2.) The unemployment rate ranges from 1% to 70% across different states.

3.)There are some states with relatively low unemployment rates, such as Andhra Pradesh, Goa, and Kerala, with rates around 1-10%.

4.)On the other hand, there are states with high unemployment rates, such as Bihar, Jharkhand, and Uttar Pradesh, with rates above 60%.

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], s=df['Estimated Unemployment Rate (%)']*100, c='red', alpha=0.5, label='Unemployment Rate')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Estimated Unemployment Rate by Location')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

From the above visualization we can conclude:

* The unemployment rate in Delhi decreased significantly over time, from 50% in the first month to 30% in the last month.

* The unemployment rate in Andhra Pradesh decreased from 60% in the first month to 70% in the last month, which is a positive trend.

* The unemployment rate in Maharashtra decreased from 50% in the first month to 20% in the last month, which is a significant improvement.

* The unemployment rate in Karnataka decreased from 30% in the first month to 10% in the last month, which is a positive trend.

* The unemployment rate in Kerala reached from 8% in the first month to 30% in the last month, which is a negative trend.

```python
import plotly.express as px
plot_unemp = df[['Estimated Unemployment Rate (%)','State']]
df_unemployed = plot_unemp.groupby('State').mean().reset_index()

df_unemployed = df_unemployed.sort_values('Estimated Unemployment Rate (%)')

fig = px.bar(df_unemployed, x='State',y='Estimated Unemployment Rate (%)',color = 'State',title = 'Average unemployment rate in each state',
             template='seaborn')
fig.show()
```

From the above graph we conclude that :

Haryana has the highest mean unemployment rate and Meghalya has the lowest

```python
fig = px.bar(df, x='Location', y='Estimated Unemployment Rate (%)', animation_frame='Month', color='State',
             title='Unemployment rate across regions from Jan. 2020 to Oct. 2020', height=700, template='seaborn')

fig.update_layout(xaxis={'categoryorder': 'total descending'})
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 1000
fig.show()
```

From above visualization We see that during the month of April, the states Puducherry, Tamil Nadu, Jharkhand, Bihar, Tripura, Haryana of India saw the major unemplyment hike and it gradually goes down in June.

```python
fig = px.scatter_geo(df,'Longitude', 'Latitude', color="Location",
                     hover_name="State", size="Estimated Unemployment Rate (%)",
                     animation_frame="Month",scope='asia',template='seaborn',title='Impact of lockdown on Employment across regions')

fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 3000

fig.update_geos(lataxis_range=[5,35], lonaxis_range=[65, 100],oceancolor="#3399FF", showocean=True)

fig.show()
```

From above visualization we conclude that he northern regions of India seems to have more unemployed people.
